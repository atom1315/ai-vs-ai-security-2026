##  РАЗДЕЛ 4 — Уязвимости / Vulnerabilities / Schwachstellen

```python
# ═══════════════════════════════════════════════════════
# CELL 8 — VULNERABILITY HEATMAP
# ═══════════════════════════════════════════════════════

vuln_models = ['Claude\nOpus 4.6', 'GPT-5.4', 'Gemini\n3.1 Pro', 'Grok 4', 'DeepSeek\nV4', 'Nova Pro']
vuln_types  = ['Prompt\nInjection', 'SSRF', 'Data\nPoisoning', 'Jailbreak', 'Info\nDisclosure', 'Model\nExtract.']

# Exploit difficulty: 1=easy to exploit, 10=hard to exploit (higher = safer)
vuln_matrix = np.array([
    [6.5, 7.0, 7.5, 6.5, 8.0, 7.0],  # Prompt Injection
    [7.0, 6.8, 7.2, 6.0, 5.5, 8.0],  # SSRF
    [5.5, 5.5, 5.8, 5.5, 4.5, 6.0],  # Data Poisoning
    [6.0, 7.5, 7.0, 5.0, 4.0, 7.5],  # Jailbreak
    [6.5, 6.5, 7.0, 5.5, 5.0, 7.8],  # Info Disclosure
    [6.0, 6.0, 6.5, 6.0, 5.5, 6.5],  # Model Extraction
])

# Custom colormap: red (vulnerable) → green (protected)
cmap = LinearSegmentedColormap.from_list(
    'rg', [(0,'#7B241C'),(0.4,'#E74C3C'),(0.6,'#F39C12'),(0.8,'#27AE60'),(1,'#1E8449')]
)

fig, ax = plt.subplots(figsize=(12, 7))
im = ax.imshow(vuln_matrix, cmap=cmap, aspect='auto', vmin=3, vmax=9)

ax.set_xticks(range(len(vuln_models)))
ax.set_xticklabels(vuln_models, fontsize=10, color=COLORS['text'])
ax.set_yticks(range(len(vuln_types)))
ax.set_yticklabels(vuln_types, fontsize=10, color=COLORS['text'])

for i in range(len(vuln_types)):
    for j in range(len(vuln_models)):
        val = vuln_matrix[i, j]
        color_txt = 'white' if val < 6.5 else COLORS['dark']
        ax.text(j, i, f'{val:.1f}', ha='center', va='center',
                fontsize=11, fontweight='bold', color=color_txt)

cbar = plt.colorbar(im, ax=ax, fraction=0.03, pad=0.02)
cbar.set_label('Exploit Difficulty\n(1=Easy, 10=Hard to Exploit)',
               color=COLORS['text'], fontsize=9)
cbar.ax.yaxis.set_tick_params(color=COLORS['text'])
plt.setp(cbar.ax.yaxis.get_ticklabels(), color=COLORS['text'])

ax.set_title(
    'VULNERABILITY HEATMAP — Exploit Difficulty by Model 2026\n'
    'Тепловая карта уязвимостей (выше = сложнее взломать) · höher = schwerer zu exploitieren',
    fontsize=12, fontweight='bold', color=COLORS['text'], pad=15
)
ax.set_xlabel('AI Model', fontsize=11, color=COLORS['text'])
ax.set_ylabel('Vulnerability Type', fontsize=11, color=COLORS['text'])

plt.tight_layout()
plt.savefig('../figures/06_vulnerability_heatmap.png')
plt.show()
print('💾 Saved: figures/06_vulnerability_heatmap.png')
print('\n🔴 Красный = легко эксплуатировать / easy to exploit / leicht zu exploitieren')
print('🟢 Зелёный = трудно эксплуатировать / hard to exploit / schwer zu exploitieren')
```

##  РАЗДЕЛ 5 — Матрица доверия / Trust Matrix / Vertrauensmatrix

```python
# ═══════════════════════════════════════════════════════
# CELL 9 — TRUST MATRIX VISUALIZATION
# ═══════════════════════════════════════════════════════

trust_data = pd.DataFrame({
    'Task_RU': [
        'Автономный анализ логов',
        'Генерация Threat Reports',
        'CVE Prioritization',
        'Code Security Review',
        'Incident Triage (с надзором)',
        'Полностью авт. SOAR Actions',
        'Автономная реконфиг. сети',
        'Autonomous Pentest в PROD',
        'ART vs ABT без надзора',
    ],
    'Task_EN': [
        'Autonomous Log Analysis',
        'Threat Report Generation',
        'CVE Prioritization',
        'Code Security Review',
        'Incident Triage (supervised)',
        'Full Autonomous SOAR Actions',
        'Autonomous Net Reconfiguration',
        'Autonomous Pentest in PROD',
        'ART vs ABT Unsupervised',
    ],
    'Trust_Pct': [92, 90, 88, 85, 78, 58, 35, 22, 15],
    'Risk_Level': ['HIGH TRUST','HIGH TRUST','HIGH TRUST','HIGH TRUST',
                   'MEDIUM','CAUTION','HIGH RISK','CRITICAL','FORBIDDEN'],
})

def risk_color(lvl):
    return {'HIGH TRUST': COLORS['green'], 'MEDIUM': COLORS['blue'],
            'CAUTION': COLORS['yellow'], 'HIGH RISK': COLORS['orange'],
            'CRITICAL': COLORS['red'], 'FORBIDDEN': '#7B241C'}[lvl]

trust_data['Color'] = trust_data['Risk_Level'].apply(risk_color)

fig, ax = plt.subplots(figsize=(14, 8))
df_sorted = trust_data.sort_values('Trust_Pct')

bars = ax.barh(
    df_sorted['Task_EN'],
    df_sorted['Trust_Pct'],
    color=df_sorted['Color'],
    height=0.65,
    edgecolor='none',
    alpha=0.88
)

# Background fills
ax.axvspan(0, 50,  alpha=0.04, color=COLORS['red'],   zorder=0)
ax.axvspan(50, 75, alpha=0.04, color=COLORS['yellow'], zorder=0)
ax.axvspan(75, 100,alpha=0.04, color=COLORS['green'],  zorder=0)

for bar, row in zip(bars, df_sorted.itertuples()):
    ax.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
            f'{row.Trust_Pct}%  [{row.Risk_Level}]',
            va='center', fontsize=9, color=row.Color, fontweight='bold')
    # RU label on bar
    ax.text(2, bar.get_y() + bar.get_height()/2,
            row.Task_RU, va='center', fontsize=8,
            color='white', alpha=0.7)

# Zone labels
ax.text(25, -0.8, 'HIGH RISK ZONE', ha='center', va='center',
        fontsize=9, color=COLORS['red'], alpha=0.6, style='italic')
ax.text(62, -0.8, 'CAUTION', ha='center', va='center',
        fontsize=9, color=COLORS['yellow'], alpha=0.6, style='italic')
ax.text(87, -0.8, 'SAFE ZONE', ha='center', va='center',
        fontsize=9, color=COLORS['green'], alpha=0.6, style='italic')

ax.set_xlim(0, 115)
ax.set_xlabel('Trust Level %', fontsize=11)
ax.set_title(
    'TRUST MATRIX — How Much to Trust AI in Infrastructure Management 2026\n'
    'Матрица доверия ИИ-инфраструктуре · KI-Vertrauensmatrix für Infrastrukturmanagement',
    fontsize=12, fontweight='bold', color=COLORS['text'], pad=15
)

# Legend
legend_items = [mpatches.Patch(color=risk_color(r), label=r)
                for r in ['HIGH TRUST','MEDIUM','CAUTION','HIGH RISK','CRITICAL','FORBIDDEN']]
ax.legend(handles=legend_items, loc='lower right',
          facecolor=COLORS['bg2'], edgecolor='#30363D',
          labelcolor=COLORS['text'], fontsize=9)

plt.tight_layout()
plt.savefig('../figures/07_trust_matrix.png')
plt.show()
print('💾 Saved: figures/07_trust_matrix.png')
```
